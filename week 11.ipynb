```python
#W11A3
def is_happy(a):
    seen = set()
    while a != 1 and a not in seen:
        seen.add(a)
        a = sum(int(ch)**2 for ch in str(a))
    return a == 1
try:
    a = int(input())
    if is_happy(a) and a > 0:
        print("YES")
    else:
        print("NO")
except ValueError:
    print("NO")
    
```
