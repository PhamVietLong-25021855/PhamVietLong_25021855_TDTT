```python
#W7A1: Tìm kiếm nhị phân 
def binary_search(arr,left,right,target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1 
arr = list(map(int,input().split()))
target = int(input())
print(binary_search(arr,0,len(arr)-1,target))
```

```python
#W7A2 cách 2 
def count_occurrences(arr, x) -> int:
    count = 0
    for value in arr:
        if value == x:
            count += 1
    return count
arr = list(map(int,input().split()))
x = int(input())
print(count_occurrences(arr,x))
```

```python
#W7A3 Sắp xếp nổi 
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] >= arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
arr = list(map(int,input().split()))
print(bubble_sort(arr))
```

```python
#W7A4
def most_frequent(arr):
    dic = {}
    for i in arr:
        dic[i] =dic.get(i,0) + 1
    max_count = 0
    result = 0
    for i in arr:
        if dic[i] > max_count:
            max_count =dic[i]
            result = i
    return f"{result} xuất hiện nhiều nhất, sớm nhất, {max_count} lần"
arr = list(map(int,input().split()))
print(most_frequent(arr))
```

```python
#W7A7
def so_sanh_do_dai(num):
    n = len(num)
    if n == 0:
        return 0
    dp = [1] * n
    for i in range(n):
        for j in range(i):
            if num[i] < num[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
n = list(map(int,input().split()))
print(so_sanh_do_dai(n))
```

```python
#W7A8
def interval(intervals, queries):
    ans = []
    for i in queries :
        Min = 1000000000
        for left, right in intervals:
            if left <= i <= right :
                Min = min(Min, right - left + 1)
        if Min == 1000000000:
            ans.append(-1)
        else:
            ans.append(Min)
    return ans 
nums = list(map(int,input().split()))
queries = list(map(int, input().split()))
intervals = []
tmp = []
for i in range(0, len(nums)):
    tmp.append(nums[i])
    if (i % 2 == 1):
        intervals.append(tmp)
        tmp = []
print(interval(intervals, queries))
```

```python
#W7A9
def so_sanh_cac_so(num):
    N = []
    for i in range(len(num)):
        count = 0
        for j in range(i + 1, len(num)):
            if num[i] > num[j]:
                count += 1
        N.append(count)
    return N
num = list(map(int,input().split()))
print(so_sanh_cac_so(num))
queries = list(map(int, input().split()))
intervals = []
tmp = []
for i in range(0, len(nums)):
    tmp.append(nums[i])
    if (i % 2 == 1):
        intervals.append(tmp)
        tmp = []
print(interval(intervals, queries))
```

```python
#W7A10
nums = list(map(int, input().split()))
target = int(input())
prefix = [0]
n = len(nums)
Max = 0
for i in range(1, n):
    prefix.append((prefix[i - 1] + nums[i - 1]))
    Max = max(Max, (prefix[i - 1] + nums[i - 1]) % target)
prefix.append(prefix[n - 1] + nums[n - 1])
if (Max == target - 1):
    print(Max)
else:
    for i in range(1, n):
        for j in range(0, i):
            if ((prefix[i] - prefix[j]) % target > Max):
                Max = (prefix[i] - prefix[j]) % target
    print(Max)
```

```python
#W7A13
nums = list(map(int, input().split()))
nums.sort()
dick = {}
n = len(nums)
Max = 0
for i in nums:
    if (i not in dick):
        dick[i] = 1
    else:
        dick[i] += 1
nums = list(set(nums))
n = len(nums)
for i in range(1, n):
    if (nums[i] - nums[i - 1] == 1):
        Max = max(Max, dick[nums[i]] + dick[nums[i - 1]])
print(Max)
```
