```python
#W6A10
n = list(map(int,input().split()))
k = int(input())
m = []
for i in range(len(n)):
    a = n[i]
    for j in range(i + 1, len(n)):
        b = n[j]
        if a + b == k:
            m.append((a,b))
res = list(set(m))
res.sort(key = lambda x:x[0])
print(res)
```

```python
#W6A16
n = int(input("Nhập vào số hàng:"))
m = int(input("Nhập vào số cột:"))
# Tạo mảng
ma_tran = []
for i in range(n):
    row = []
    for j in range(m):
        value = int(input(f"a[{i}][{j}]"))
        row.append(value)
    ma_tran.append(row)
for row in ma_tran:
    for i in row:
        print(f"{i:>4}",end = "")
    print()
```

```python
#W6A18
n = int(input("Nhập vào số hàng:"))
m = int(input("Nhập vào số cột:"))
k = int(input("Nhập vào cột thứ k:")) - 1
ma_tran = []
for i in range(n):
    row = []
    for j in range(m):
        value = int(input(f"a[{i}][{j}]"))
        row.append(value)
    ma_tran.append(row)
if 0 <= k <= m:
    tong_cot = 0
    for i in range(n):
        tong_cot += ma_tran[i][k]
        print(f"{ma_tran[i][k]:>4}", end ="")
    print(f" :là cột thứ {k + 1}")
    print(f"Tổng của cột {k + 1}:",tong_cot)
```
