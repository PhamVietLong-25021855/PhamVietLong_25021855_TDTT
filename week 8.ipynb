```python
#W8A1
class cylinder:
    def __init__(self,r,h):
        self.r = r
        self.h = h
        self.pi = 3.14
    def surface_arae(self):
        return 2*self.pi*self.r*(self.r + self.h)
    def vol(self):
        return self.pi*self.r*self.r*self.h
r,h = map(float,input().split())
Cyl = cylinder(r,h)
are = Cyl.surface_arae()
vole = Cyl.vol()
print(f"{are:.2f},{vole:.2f}")
```

```python
#W8A2
class Date:
    def __init__(self,dd,mm,yyyy):
        self.dd = dd
        self.mm = mm
        self.yyyy = yyyy
    def years(self):
        return (self.yyyy % 4 == 0 and self.yyyy % 100 != 0) or self.yyyy % 400 == 0
    def days(self):
        if self.mm in [1,3,5,7,8,10,12]:
                return 31
        elif self.mm in [4,6,9,11]:
                return 30
        elif self.mm == 2:
            if self.years():
                return 29
            else:
                return 28
        return 0
    def is_valid(self):
        if self.yyyy <= 0 or self.mm <1 or self.mm > 12:
            return False
        if self.dd < 1 or self.dd > self.days():
            return False
        return True 
    def next_day(self):
        day = self.dd + 1
        month = self.mm
        years = self.yyyy
        if day > self.days():
            day = 1
            month += 1
            if month > 12:
                month = 1 
                years += 1
        return Date(day,month,years)
dd,mm,yyyy = map(int,input().split("/"))
dat = Date(dd,mm,yyyy)
if not dat.is_valid():
    print("INVALID")
else:
    next_dy = dat.next_day()
    print(f"{next_dy.dd:02d}/{next_dy.mm:02d}/{next_dy.yyyy:04d}")
```

```python
#W8A3
class coordinates:
    def __init__ (self,x,y):
        self.x = x
        self.y = y
    def check(self):
        if self.x == self.y == 0:
            return "Là gốc tọa độ"
        elif self.x == 0 and self.y != 0:
            return f"Thuộc trục Oy và khoảng cách = {abs(self.y):.2f}"
        elif self.x != 0 and self.y == 0:
            return f"Thuộc trục Ox và khoảng cách = {abs(self.x):.2f}"
        else:
            return f"Khoảng cách = {((self.x)**2 + (self.y)**2)**(1/2):.2f}"
x,y = map(int,input().split())
coor = coordinates(x,y)
khoang_cach = coor.check()
print(khoang_cach)
        
```

```python
#W8A4
class caculater:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a*self.b
    def divide(self):
        if self.b == 0:
            return False
        else:
            return self.a/self.b
    def power(self):
        return self.a**self.b
    def mod(self):
        return self.a % self.b
    def set_number(self,a,b):
        self.a = a
        self.b = b
        return caculater(a,b)
a,b = map(float,input("Nhập vào các giá trị của a và b").split())
cal = caculater(a,b)
while True:
    choice = input("Nhập vào chức năng mà bạn muốn chọn:")
    print("-- Các chức năng của máy tính--")
    print("1.Cộng")
    print("2.Trừ")
    print("3.Nhân")
    print("4.Chia")
    print("5.Hàm mũ")
    print("6.Lấy phần dư")
    print("7.Cập nhật 2 số mới")
    if choice == '1':
        print("Kết quả:",cal.add())
    elif choice == '2':
        print("Kết quả:",cal.subtract())
    elif choice == '3':
        print("Kết quả:",cal.product())
    elif choice == '4':
        print("Kết quả:",cal.divide())
    elif choice == '5':
        print("Kết quả:",cal.power())
    elif choice == '6':
        print("Kết quả:",cal.mod())
    elif choice == '7':
        a,b = map(float,input("Nhập vào các giá trị của a và b").split())
        print("Nhập lại số mới mà bạn muốn chọn:",cal.set_number(a,b))
    else:
        print("Không hợp lệ. Vui lòng thử lại")
    exit_choice = input("Bạn có muốn thoát")
    if exit_choice == 'yes' or exit_choice == 'YES' or exit_choice == 'Yes':
        break
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.items = []
    def add(self,name,price):
        self.items.append({"name":name, "price":price})
    def removed(self,name):
        for items in self.items:
            if items["name"] == name:
                self.items.remove(items)
                return True
        return False
    def is_empty(self):
        if len(self.items) == 0:
            return 0
    def total_price(self):
        total = 0
        for items in self.items:
            total += items["price"]
        return total
    def display_items(self):
        if self.is_empty():
            print("Giỏ hàng rỗng")
        else:
            for item in self.items:
                print(f"Tên: {item['name']}, Giá: {item['price']}")
    def remove_all(self):
        for item in self.items:
            self.items.remove(item)
        return self.items
cart = ShoppingCart()
while True:
    print("\n----- MENU -----")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng có rỗng không")
    print("5. Tính tổng tiền")
    print("6. Hiển thị toàn bộ sản phẩm")
    print("7. Xóa toàn bộ giỏ hàng")
    choice = input("Chọn chức năng (2-7): ")
    if choice == "2":
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        cart.add(name, price)
    elif choice == "3":
        name = input("Nhập tên sản phẩm cần xóa: ")
        if cart.removed(name):
            print("Đã xóa sản phẩm")
        else:
            print("Không tìm thấy sản phẩm")
    elif choice == "4":
        if cart.is_empty():
            print("Giỏ hàng đang rỗng")
        else:
            print("Giỏ hàng không rỗng")
    elif choice == "5":
        print("Tổng tiền:", cart.total_price())
    elif choice == "6":
        cart.display_items()
    elif choice == "7":
        cart.remove_all()
        print("Đã xóa toàn bộ giỏ hàng")
    else:
        print("Lựa chọn không hợp lệ")
    exit_choice = input("Có muốn thoát khỏi chương trình không? ")
    if exit_choice == "yes" or exit_choice == "Yes" or exit_choice == "YES":
        break
```

```python
#W8A6
class Rectangle:
    def __init__(self,w,h,k):
        self.w = w
        self.h = h
        self.k = k
    def area(self):
        if self.w == 0 or self.h == 0 or self.k == 0:
            return False
        return f"{self.w*self.h*self.k*self.k:.2f}"
    def perimeter(self):
        if self.w == 0 or self.h == 0 or self.k == 0:
            return False
        return f"{self.k*(self.w + self.h)*2:.2f}"
w,h,k = map(float,input().split())
rec = Rectangle(w,h,k)
print(rec.area(),rec.perimeter())
```

```python
#W8A7
import math
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không được bằng 0")
        if den < 0:
            num = -num
            den = -den
        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g
    def add(self, other):
        num = self.num * other.den + other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)
    def sub(self, other):
        num = self.num * other.den - other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)
    def mul(self, other):
        num = self.num * other.num
        den = self.den * other.den
        return Fraction(num, den)
    def div(self, other):
        num = self.num * other.den
        den = self.den * other.num
        return Fraction(num, den)
    def __str__(self):
        return f"{self.num}/{self.den}"
a, b, op, c, d = input().split()
a, b, c, d = int(a), int(b), int(c), int(d)
f1 = Fraction(a, b)
f2 = Fraction(c, d)
if op == '+':
    result = f1.add(f2)
elif op == '-':
    result = f1.sub(f2)
elif op == '*':
    result = f1.mul(f2)
elif op == '/':
    result = f1.div(f2)
print(result)
```

```python
#W8A8
class BankAccount:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
    def deposit(self,x):
        self.balance += x
        return self.balance
    def withdraw(self,x):
        self.balance -= x
        return self.balance
    def __str__(self):
        return f"{self.owner}, {self.balance:.2f}"
onwer, initial_balance = input().split()
initial_balance = float(initial_balance)
n = int(input())
account = BankAccount(onwer,initial_balance)
for _ in range(n):
    comman, x = input().split()
    x = float(x)
    if comman == "DEPOSIT":
        account.deposit(x)
    elif comman == "WITHDRAW":
        account.withdraw(x)
print(f"{account.balance:.2f}")
```

```python
#W8A9
class Students:
    def __init__(self,name):
        self.name = name
        self.score = {}
    def add_score(self,subject, score):
        self.score[subject] = score 
    def avg(self):
        if not self.score:
            return 0
        else:
            m = 0
            for key,value in self.score.items():
                m += float(value)
        return m/len(self.score)
    def rank(self):
        if self.avg() >= 8:
            return "Excellent"
        elif 8 > self.avg() >= 6.5:
            return "Good"
        elif 6.5 > self.avg() >= 5:
            return "Average"
        else:
            return "Poor"
name = input().strip()
n = int(input())
student = Students(name)
for _ in range(n):
    subject,score = input().split()
    student.add_score(subject,float(score))
average = student.avg()
ranking = student.rank()
print(f"{student.name} {average:.2f} {ranking}")   
```

```python
#W8A10
class Book:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.library = []
    def add(self,book):
        return self.library.append(book)
    def find_count_author(self,author):
        count = 0
        for i in self.library:
            if i.author == author:
                count += 1
        return count
    def find_countyear_year(self,year):
        count = 0 
        for i in self.library:
            if i.year == year:
                count += 1
        return count
library = Library()
n = int(input())
for _ in range(n):
    line = input().strip()
    if line.startswith("ADD"):
        _, data = line.split(" ", 1)
        title, author, year = data.split(";")
        book = Book(title, author, int(year))
        library.add(book)
    elif line.startswith("COUNTYEAR"):
        _,year = line.split()
        print(library.find_countyear_year(int(year)))
    elif line.startswith("COUNT"):
        _,author = line.split(" ",1)
        print(library.find_count_author(author))
```

```python
#W8A11
import math
def creat_vector(a,b):
    return [a[i] - b[i] for i in range(3)]
def cross(u,v):
    return [u[1]*v[2] -u[2]* v[1],
            u[2]*v[0] -u[0]* v[2],
            u[0]*v[1] - u[1]*v[0]]
def dot(u,v):
    return sum(u[i]*v[i] for i in range(3))
def norm(v):
    return math.sqrt(dot(v,v))
A = list(map(float,input().split()))
B = list(map(float,input().split()))
C = list(map(float,input().split()))
D = list(map(float,input().split()))
# Tạo vecto cho mặt phẳng ABC
AB = creat_vector(B,A)
AC = creat_vector(C,A)
n1 = cross(AB,AC)
# Tạo vecto cho mặt phẳng BCD
BC = creat_vector(C,B)
BD = creat_vector(D,B)
n2 = cross(BC,BD)
# Tính góc của 2 mặt phẳng 
cos_phi = abs(dot(n1,n2)) / (norm(n1)*norm(n2))
phi = math.degrees(math.acos(cos_phi))
print(f"{phi:.2f}")
```

```python
#W8A12
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = realPart
        self.imaginaryPart = imaginaryPart
    def muintilself(self,other):
        real = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        imaginary = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(real,imaginary)
r1, i1 = map(float,input().split())
r2, i2 = map(float,input().split())
c1 = ComplexNumber(r1, i1)
c2 = ComplexNumber(r2, i2)
result = c1.muintilself(c2)
print(f"{int(result.realPart) if result.realPart.is_integer() else result.realPart} "
      f"{int(result.imaginaryPart) if result.imaginaryPart.is_integer() else result.imaginaryPart}")
```

```python
#W8A13
class Triagle:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def check(self):
        import math
        if self.a + self.b > self.c and self.b + self.c > self.a and self.a + self.a + self.c > self.b:
            half_perimeter = (self.a + self.b + self.c)/2
            acreage = math.sqrt(half_perimeter*(half_perimeter - self.a)*(half_perimeter - self.b)*(half_perimeter - self.c))
            return f"{acreage:.2f}"
        return "invalid"
a,b,c = map(int,input().split())
kk = Triagle(a,b,c)
print(kk.check())  
```
