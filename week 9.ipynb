```python
# W9A1
def bullet_sort(a):
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            if a[i]>=a[j]:
                a[i],a[j] = a[j],a[i]
    return a
def merge(a,b):
    if len(a) == 0:
        return b
    if len(b) == 0:
        return a
    if a[0] < b[0]:
        return [a[0]] + merge(a[1:],b)
    else:
        return [b[0]] + merge(a,b[1:])
lst1 = list(map(int,input().split()))
lst2 = list(map(int,input().split()))
a = bullet_sort(lst1)
b = bullet_sort(lst2)
result = merge(a,b)
print(result)
```

```python
# W9A2
def merge(a,b):
    counta = 0
    countb = 0
    re=[]
    while True:
            try:
              va = a[counta]
            except IndexError:
              re.extend(a[counta::])
              break
            try:
                vb=b[countb]
            except IndexError:
              re.extend(a[counta::])
              break
            if va<vb:
                re.append(va)
                counta += 1

            elif vb<va:
                re.append(vb)
                countb += 1
            else:
                re.append(va)
                counta += 1
    return re
n = int(input())
k = int(input())
for _ in range(n):
    m = list(map(int,input().split()))
    if _ == 0:
        c = m
        continue
    c = merge(m,c)
print(c[k-1])
```

```python
#W9A5
def is_contiguous_sublist(list1, list2):
    m = len(list1)
    n = len(list2)
    if m == 0:
        return True
    elif m > n:
        return False
    for i in range(n - m + 1):
        ok = True
        for j in range(m):
            if list2[i+j] != list1[j]:
                ok = False
                break
        if ok :
            return True 
    return False
list1, list2 = list(map(int,input().split())), list(map(int,input().split()))
result = is_contiguous_sublist(list1, list2)
if result == True:
    print("YES")
else:
    print("NO")
```

```python
#W9A6
def w9a6(a):
    n = len(a)
    i = n - 2
    while i >= 0 and a[i] >= a[i + 1]:
        i -= 1

    if i < 0:
        a.reverse()
        return a

    j = n - 1
    while a[j] <= a[i]:
        j -= 1
    a[i], a[j] = a[j], a[i]
    a[i + 1:] = reversed(a[i + 1:])
    return a

n = int(input())
a = list(map(int, input().split()))
res = w9a6(a)

print(*res) #unpacking
```

```python
#W9A7
def canPlaceFlowers(flowerbed,k):
    count=0
    if len(flowerbed) <=1:
        if flowerbed[0] ==0:
            count+=1
    else:
        if flowerbed[0]==0 and flowerbed[1]==0:
            flowerbed[0]=1
            count+=1
        if flowerbed[len(flowerbed)-1]==0 and flowerbed[len(flowerbed)-2]==0:
            flowerbed[len(flowerbed)-1]=1
            count+=1
        for i in range(1,len(flowerbed)-1):
            if flowerbed[i-1] ==0 and flowerbed[i+1]==0 and flowerbed[i]==0:
                count+=1
    if count>=k:
        return True
    return False 

a=list(map(int,input().split()))
k=int(input())
if canPlaceFlowers(a,k):
    print("True")
else:
    print("False")
```

```python
#W9A8
def is_path_crossing(moves):
    x = y = 0
    seen = {(0,0)}
    for ch in moves:
        if ch == 'R':
            x += 1
        elif ch == 'L':
            x -= 1
        elif ch == 'U':
            y += 1
        elif ch == 'D':
            y -= 1
        else:
            continue
    if (x,y) in seen:
        return True
    seen.add((x,y))
    return False 
n = input()
result = is_path_crossing(n)
print(result)
        
```

```python
# W9A10
def lower_bound_recursive(arr,x,left,right):
    if left > right:
        return left
    mid = (left + right)//2
    if arr[mid] >= x:
        return lower_bound_recursive(arr,x,left,mid - 1)
    else:
        return lower_bound_recursive(arr,x,mid + 1,right)
def findRadius(houses,heaters):
    houses.sort()
    heaters.sort()
    r = 0
    m = len(heaters)
    for h in houses:
        idx = lower_bound_recursive(heaters,h,0,m-1)
        right = abs(heaters[idx] - h) if idx < m else float("inf")
        left = abs(h - heaters[idx - 1]) if idx > 0 else float("inf")
        r = max(r, min(left,right))
    return r
houses = list(map(int,input().split()))
heaters = list(map(int,input().split()))
print(findRadius(houses,heaters))
```
