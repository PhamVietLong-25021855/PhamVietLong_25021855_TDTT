```python

# W11A4
def analgram(path1: str, path2: str) -> bool:
    file1, file2 = open(path1, "r"), open(path2, "r")
    ans = set(''.join(file1.readline().split()).lower()) == set(''.join(file2.readline().split()).lower())
    file1.close()
    file2.close()
    return ans
print(analgram("text.txt", "text2.txt"))
```

```python
#W11A10
def inp(path):
    file = open(path, "r")
    houses = list(map(int, file.readline().strip().split()))
    file.close()
    return houses
def robber(houses):
    n = len(houses)
    house1 = 0
    house2 = 0
    for i in range(0, n):
        Max = max(houses[i] + house1, house2)
        house1 = house2
        house2 = Max
    return house2
houses = inp("text6.23")
print(robber(houses))
```

```python
#W11A12
def maximumproduct(path):
    file = open(path, "r")
    nums = list(map(int, file.readline().strip().split()))
    nums.sort()
    return max(nums[-1] * nums[-2] * nums[-3], nums[0] * nums[1] * nums[-1])
print(maximumproduct("text7.txt"))
```

```python
#W11A14
def avgtime(path):
    total = 0
    cnt = 0
    file = open(path, "r")
    lines = file.readlines()
    for line in lines[1:]:
        time = line.strip().split(",")
        total += float(time[1])
        cnt += 1
    file.close()
    return f"{total / cnt:.2f}"
print(avgtime("test.csv"))
```

```python
#W11A15
def find_movies(path):
    check = False
    movie = []
    file = open(path, "r")
    lines = file.readlines()
    for line in lines[1:]:
        des = line.strip().split(",")
        if (float(des[3]) >= 8.0 and des[2].lower() != "boring"):
            movie.append(des[1])
            check = True
    file.close()
    if (check == True):
        return movie
    else:
        return "All are boring"
print(*find_movies("text15.txt"))
```

```python
#W11A16
def perimeter(path):
    file = open(path, "r")
    shape = file.readline().strip().split(":")[1].strip()
    info = file.readline().strip()
    file.close()
    if shape == "SQUARE":
        side = float(info.split(": ")[1])
        p = 4 * side
    elif shape == "RECTANGLE":
        parts = info.split()
        width = float(parts[0].split(": ")[1])
        height = float(parts[1].split(": ")[1])
        p = 2 * (width + height)
    elif shape == "CIRCLE":
        radius = float(info.split(": ")[1])
        p = 2 * 3.14 * radius
    print(f"{p:.2f}")
perimeter("text16.txt")
```

```python
#W11A17
def productExceptSelf(path):
    file = open(path, "r")
    n = int(file.readline().strip())
    nums = list(map(int, file.readline().strip().split()))
    file.close()
    res = []
    for i in range(n):
        prod = 1
        for j in range(n):
            if i != j:
                prod *= nums[j]
        res.append(prod)
    for x in res:
        print(x, end=" ")
productExceptSelf("text17.txt")
```

```python
#W11A20
def image_smoother(path):
    file = open(path, "r")
    image = []
    lines = file.readlines()
    for line in lines:
        tmp = list(map(int, line.strip().split()))
        image.append(tmp)
    n = len(tmp)
    ans = []
    move_i = [0, 0, -1, 1, 1, -1, -1, 1]
    move_j = [1, -1, 0, 0, 1, -1, 1, -1]
    for i in range(0, n):
        tmp = []
        for j in range(0, n):
            avg = image[i][j]
            cnt = 1
            for k in range(0, 8):
                i1 = i + move_i[k]
                j1 = j + move_j[k]
                if (0 <= i1 < n and 0 <= j1 < n):
                    avg += image[i1][j1]
                    cnt += 1
            avg //= cnt
            tmp.append(avg)
        ans.append(tmp)
    return ans, n
ans, n = image_smoother("text20.txt")
for i in range(0, n):
    for j in range(0, n):
        print(ans[i][j], end = " ")
    print()
```
