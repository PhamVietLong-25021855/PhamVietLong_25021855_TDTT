```
#W5A3
def check(a):
    for i in range(2, int(a**0.5)+ 1):
        if a % i == 0:
            return False
    return True
a = int(input())
m = check(a)
print(m)
```

```
#W5A4
def sohoanhao(a):
    m = 0
    for i in range(1, a):
        if a % i == 0:
            m += i
    return m
a = int(input())
b = sohoanhao(a)
if a == b:
    print(True)
else:
    print(False)
```

```
#W5A7
num1, operat, num2 = map(str, input().split())
if operat == "-":
    k = int(num1) - int(num2)
    print(f"{k:.2f}")
elif operat == "+":
    k = int(num1) + int(num2)
    print(f"{k:.2f}")
elif operat == "/":
    k = int(num1) / int(num2)
    if int(num2) == 0:
        print("Lỗi")
    else:
         print(f"{k:.2f}")
elif operat == "*":
    k = int(num1) * int(num2)
    print(f"{k:.2f}")
```

```
#W5A8
a, b = map(int,input().split())
count = 0
while a > 0 or b > 0:
    if a % 2 != b % 2:
        count += 1
    a = a // 2
    b = b // 2
print(count)
```

```
#W5A9
n = int(input())
m = 0
while n > 0:
    k = n % 10
    m += k
    n //= 10
print(m)
```

```
#W5A10
def kt(a, b):
    if len(a) != len(b):
        return False
    map_a = {}
    map_b = {}
    for c1, c2 in zip(a, b):
        if c1 in map_b and map_b[c1] != c2:
            return False
        if c2 in map_a and map_a[c2] != c1:
            return False
    map_b[c1] = c2
    map_a[c2] = c1
    return True
a, b =map(str,input().split())
print(kt(a,b))
```
