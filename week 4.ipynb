```
# bai 7
def largest_prime_factor(n):
    i = 2
    largest = 1
    while i * i <= n:
        if n % i == 0:
            largest = i
            while n % i == 0:
                n //= i
        i += 1
    if n > 1:
        largest = n
    return int(largest)

# Nhập và in kết quả
n = int(input("Nhập số nguyên dương n (n ≥ 2): "))
print("Ước số nguyên tố lớn nhất của n là:", largest_prime_factor(n))
```

```
# bai 8
def reverse_number(n):
    return int(str(n)[::-1])

def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

def reverse_add_plaindrome(n):
    steps=0
    while not is_palindrome(n):
        n=n + reverse_number(n)
        steps +=1
    return steps , n
n=int(input("Nhập vào một số nguyên dương n:"))
steps, palindrome=reverse_add_plaindrome(n)
print(steps)
print(palindrome)
```

```
#bai 9
import math

def perfect_squares_distinct_digits(n):
    if n <= 0:

        print("(no number)")
        return
    limit = int(math.sqrt(n))
    results = []
    def has_distinct_digits(number):
        s = str(number)
        distinct_digits = set(s)
        return len(s) == len(distinct_digits)
    for i in range(1, limit + 1):
        square = i * i
        if has_distinct_digits(square):
            results.append(square)
    if results:
        print(" ".join(map(str, results)))
    else:
       print("(no number)")
try:
    n_input = input("Nhập vào số nguyên dương n: ").strip()
    n = int(n_input)

    if n <= 0:
        print("Lỗi: n phải là số nguyên dương.")
    else:
        perfect_squares_distinct_digits(n)

except ValueError:
    print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")

```

```
# bai 10
def solve_longest_collatz(n):
    if n < 1:
        return "Lỗi: n không phải là số nguyên dương"
    collatz_length={1:1}
    max_L = 1
    result_x= 1
    def get_collatz_length(i):
        if i in collatz_length :
            return collatz_length[i]
        current_i=i
        path=[]
        while current_i not in collatz_length:
            path.append(current_i)
            if current_i % 2 ==0:
                current_i = current_i // 2
            else:
                current_i=3 * current_i + 1
        length_at_stop=collatz_length[current_i]
        total_length = length_at_stop
        for num in reversed(path):
            total_length += 1
            collatz_length[num] = total_length
        return collatz_length[i]
    for x in range(2, n + 1):
        L_x = get_collatz_length(x)
        if L_x > max_L:
            max_L = L_x
            result_x =x
    print(f"{result_x} {max_L}")
try:
    n_input = input("Nhập vào số nguyên dương n:")
    n = int(n_input)

    if n<= 0:
        print("n không phải là số nguyên dương")
    else:
        solve_longest_collatz(n)
except ValueError:
    print("Đầu vào không hợp lệ ")

```

```
# bài 11
import math

def count_even_divisors(n):

    if n <= 0:
        return 0
    temp_n = n
    k = 0
    while temp_n > 0 and temp_n % 2 == 0:
        k += 1
        temp_n //= 2
    if k == 0:
         return 0
    count_even = k
    i = 3
    while i * i <= temp_n:
        if temp_n % i == 0:
            a_i = 0
            while temp_n % i == 0:
                a_i += 1
                temp_n //= i
            count_even *= (a_i + 1)
        i += 2
    if temp_n > 1:
        count_even *= 2

    return count_even
try:
    n_input = input(f"Nhập vào số nguyên dương n (< {10**6}): ").strip()
    n = int(n_input)

    if n <= 0 or n >= 10**6:
        print("Lỗi: n phải là số nguyên dương và nhỏ hơn 10^6.")
    else:
        result = count_even_divisors(n)
        print(f"Số lượng ước số chẵn của {n} là: {result}")

except ValueError:
    print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
```

```
#Bài 13
def sum_of_divisors(n):
    """Tính tổng các ước của n (trừ chính nó)"""
    total = 1  # 1 luôn là ước của mọi số > 1
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
    return total if n > 1 else 0

# Nhập hai số
a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

# Kiểm tra điều kiện thân thiết
if sum_of_divisors(a) == b and sum_of_divisors(b) == a:
    print("true")
else:
    print("false")
```

```
# bài 14
def find_gcd(m, n):
    if m <= 0 or n <= 0:
        raise ValueError("Cả hai số phải là số nguyên dương.")
    while n:
        m, n = n, m % n
    return m
try:
    m_input = input("Nhập số nguyên dương m: ").strip()
    n_input = input("Nhập số nguyên dương n: ").strip()

    m = int(m_input)
    n = int(n_input)

    if m <= 0 or n <= 0:
        print("Lỗi: Cả hai số phải là số nguyên dương.")
    else:
        result = find_gcd(m, n)
        print(f"Ước chung lớn nhất (GCD) của {m} và {n} là: {result}")

except ValueError as e:
    if "Cả hai số phải là số nguyên dương." in str(e):
        print(e)
    else:
        print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập số nguyên dương.")
```

```
# bai 15
def chicken_and_dog_solver():
    """
    Giải bài toán 'Vừa gà vừa chó' dựa trên tổng số con và tổng số chân.
    """
    try:
        total_animals_str = input("Nhập tổng số con (C): ").strip()
        C = int(total_animals_str)
        total_legs_str = input("Nhập tổng số chân (L): ").strip()
        L = int(total_legs_str)
        if C <= 0 or L <= 0:
            print("invalid")
            return
        if L % 2 != 0:
            print("invalid")
            return
        if L < 2 * C or L > 4 * C:
            print("invalid")
            return
        D = (L - 2 * C) // 2
        G = C - D
        print(f"Số lượng gà: {G}")
        print(f"Số lượng chó: {D}")

    except ValueError:
        print("invalid")
    except Exception:
        print("invalid")
def example_case(C, L):
    """
    Hàm test với giá trị cố định và in ra kết quả cuối cùng theo format yêu cầu.
    """
    try:
        if C <= 0 or L <= 0:
            print("invalid")
            return

        if L % 2 != 0 or L < 2 * C or L > 4 * C:
            print("invalid")
            return

        D = (L - 2 * C) // 2
        G = C - D
        print(f"{G} {D}")

    except Exception:
        print("invalid")
print("--- Test case từ đề bài (36 con, 100 chân) ---")
example_case(36, 100) # G = 22, D = 14
```

```
# bài 27
s = input("Nhập chuỗi: ")

def str_len(t):
    c = 0
    for _ in t: c += 1
    return c

def count_words(t):
    n = str_len(t)
    i = 0
    cnt = 0
    while i < n:
        while i < n and t[i] == ' ':
            i += 1
        if i < n:
            cnt += 1
            while i < n and t[i] != ' ':
                i += 1
    return cnt

print("Số từ:", count_words(s))
```

```
# bài 28
s = input("Nhập chuỗi: ")

def first_word(t):
    # bỏ khoảng trắng đầu
    i = 0
    for ch in t:
        if ch != ' ': break
        i += 1
    # lấy đến trước khoảng trắng tiếp theo
    kq = ""
    j = i
    while j < i + sum(1 for _ in t[i:]) and t[j] != ' ':
        kq += t[j]
        j += 1
    return kq

print("Từ đầu tiên:", first_word(s))

```

```
# bài 33
n = int(input("Nhập số nguyên: "))

neg = n < 0
n = abs(n)
s = str(n)

# chèn dấu chấm từ phải sang trái
res = ""
cnt = 0
for ch in s[::-1]:
    if cnt == 3:
        res += '.'
        cnt = 0
    res += ch
    cnt += 1
res = res[::-1]
if neg: res = "-" + res

print("Kết quả:", res)
```
